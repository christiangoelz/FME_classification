```python
import pickle 
import pandas as pd 
from pathlib import Path
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 

folder = Path.cwd().parent / 'results' / 'group_classification'

# Force feature 
with open(str(folder / 'force' / 'performance' / 'all' / 'svc'/ 'confmat.pkl'), 'rb') as handle:
    confmat_force = pickle.load(handle) 
with open(str(folder / 'force' / 'performance' / 'steady_left' / 'svc'/ 'confmat.pkl'), 'rb') as handle:
    stl_force = pickle.load(handle)
with open(str(folder / 'force' / 'performance' / 'sine_left' / 'svc'/ 'confmat.pkl'), 'rb') as handle:
    sl_force = pickle.load(handle)    
with open(str(folder / 'force' / 'performance' / 'steady_right' / 'svc'/ 'confmat.pkl'), 'rb') as handle:
    str_force = pickle.load(handle)
with open(str(folder / 'force' / 'performance' / 'sine_right' / 'svc'/ 'confmat.pkl'), 'rb') as handle:
    sr_force = pickle.load(handle)     

# EEG feature
with open(str(folder / 'eeg' / 'dmd' / 'all' / 'svc'/ 'confmat.pkl'), 'rb') as handle:
    confmat_eeg = pickle.load(handle)
with open(str(folder / 'eeg' / 'dmd' / 'steady_left' / 'svc'/ 'confmat.pkl'), 'rb') as handle:
    stl_eeg = pickle.load(handle)
with open(str(folder / 'eeg' / 'dmd' / 'sine_left' / 'svc'/ 'confmat.pkl'), 'rb') as handle:
    sl_eeg = pickle.load(handle)    
with open(str(folder / 'eeg' / 'dmd' / 'steady_right' / 'svc'/ 'confmat.pkl'), 'rb') as handle:
    str_eeg = pickle.load(handle)
with open(str(folder / 'eeg' / 'dmd' / 'sine_right' / 'svc'/ 'confmat.pkl'), 'rb') as handle:
    sr_eeg = pickle.load(handle) 

pal = cmap = sns.color_palette(0, 20, as_cmap=True)
sns.set_context('paper', font_scale = .9)
labels = ["NOV", "FME"]
cm = 1/2.54
fig, axes = plt.subplots(2,5, figsize = (19*cm, 7*cm))

#eeg
conf_mat_eeg = pd.DataFrame(np.mean(confmat_eeg, axis = 0), columns = labels, index = labels)
conf_mat_eeg.index.name = 'Actual class'
conf_mat_eeg.columns.name = 'Predicted class'
sns.heatmap(conf_mat_eeg, cmap= 'Blues', annot=True, fmt='.2', vmin = 0, vmax = 1, ax = axes[0,0])
axes[0,0].set_xticklabels([])
axes[0,0].set_xlabel('')
axes[0,0].set_title('all tasks')
axes[0,0].text(-1, -0.25, "a", horizontalalignment='left', size='medium', color='black', weight='semibold')

conf_mat_eeg = pd.DataFrame(np.mean(stl_eeg, axis = 0), columns = labels, index = labels)
conf_mat_eeg.index.name = 'Actual class'
conf_mat_eeg.columns.name = 'Predicted class'
sns.heatmap(conf_mat_eeg, cmap= 'Blues', annot=True, fmt='.2', vmin = 0, vmax = 1, ax = axes[0,1])
axes[0,1].set_xticklabels([])
axes[0,1].set_xlabel('')
axes[0,1].set_yticklabels([])
axes[0,1].set_ylabel('')
axes[0,1].set_title('steady left')

conf_mat_eeg = pd.DataFrame(np.mean(sl_eeg, axis = 0), columns = labels, index = labels)
conf_mat_eeg.index.name = 'Actual class'
conf_mat_eeg.columns.name = 'Predicted class'
sns.heatmap(conf_mat_eeg, cmap= 'Blues', annot=True, fmt='.2', vmin = 0, vmax = 1, ax = axes[0,2])
axes[0,2].set_xticklabels([])
axes[0,2].set_xlabel('')
axes[0,2].set_yticklabels([])
axes[0,2].set_ylabel('')
axes[0,2].set_title('sine left')

conf_mat_eeg = pd.DataFrame(np.mean(str_eeg, axis = 0), columns = labels, index = labels)
conf_mat_eeg.index.name = 'Actual class'
conf_mat_eeg.columns.name = 'Predicted class'
sns.heatmap(conf_mat_eeg, cmap= 'Blues', annot=True, fmt='.2', vmin = 0, vmax = 1, ax = axes[0,3])
axes[0,3].set_xticklabels([])
axes[0,3].set_xlabel('')
axes[0,3].set_yticklabels([])
axes[0,3].set_ylabel('')
axes[0,3].set_title('steady right')

conf_mat_eeg = pd.DataFrame(np.mean(sr_eeg, axis = 0), columns = labels, index = labels)
conf_mat_eeg.index.name = 'Actual class'
conf_mat_eeg.columns.name = 'Predicted class'
sns.heatmap(conf_mat_eeg, cmap= 'Blues', annot=True, fmt='.2', vmin = 0, vmax = 1, ax = axes[0,4])
axes[0,4].set_xticklabels([])
axes[0,4].set_xlabel('')
axes[0,4].set_yticklabels([])
axes[0,4].set_ylabel('')
axes[0,4].set_title('sine right')

#force
conf_mat_force = pd.DataFrame(np.mean(confmat_force,axis = 0), columns = labels, index = labels)
conf_mat_force.index.name = 'Actual class'
conf_mat_force.columns.name = 'Predicted class'
sns.heatmap(conf_mat_force, cmap= 'Blues', annot=True, fmt='.2', vmin = 0, vmax = 1, ax = axes[1,0])

conf_mat_force = pd.DataFrame(np.mean(stl_force, axis = 0), columns = labels, index = labels)
conf_mat_force.index.name = 'Actual class'
conf_mat_force.columns.name = 'Predicted class'
sns.heatmap(conf_mat_force, cmap= 'Blues', annot=True, fmt='.2', vmin = 0, vmax = 1, ax = axes[1,1])
axes[1,1].set_xticklabels([])
axes[1,1].set_xlabel('')
axes[1,1].set_yticklabels([])
axes[1,1].set_ylabel('')


conf_mat_force = pd.DataFrame(np.mean(sl_force, axis = 0), columns = labels, index = labels)
conf_mat_force.index.name = 'Actual class'
conf_mat_force.columns.name = 'Predicted class'
sns.heatmap(conf_mat_force, cmap= 'Blues', annot=True, fmt='.2', vmin = 0, vmax = 1, ax = axes[1,2])
axes[1,2].set_xticklabels([])
axes[1,2].set_xlabel('')
axes[1,2].set_yticklabels([])
axes[1,2].set_ylabel('')


conf_mat_force = pd.DataFrame(np.mean(str_force, axis = 0), columns = labels, index = labels)
conf_mat_force.index.name = 'Actual class'
conf_mat_force.columns.name = 'Predicted class'
sns.heatmap(conf_mat_force, cmap= 'Blues', annot=True, fmt='.2', vmin = 0, vmax = 1, ax = axes[1,3])
axes[1,3].set_xticklabels([])
axes[1,3].set_xlabel('')
axes[1,3].set_yticklabels([])
axes[1,3].set_ylabel('')


conf_mat_force = pd.DataFrame(np.mean(sr_force, axis = 0), columns = labels, index = labels)
conf_mat_force.index.name = 'Actual class'
conf_mat_force.columns.name = 'Predicted class'
sns.heatmap(conf_mat_force, cmap= 'Blues', annot=True, fmt='.2', vmin = 0, vmax = 1, ax = axes[1,4])
axes[1,4].set_xticklabels([])
axes[1,4].set_xlabel('')
axes[1,4].set_yticklabels([])
axes[1,4].set_ylabel('')

axes[1,0].text(-1, -0.25, "b", horizontalalignment='left', size='medium', color='black', weight='semibold')
fig.tight_layout()
fig.savefig(str(Path.cwd().parent / 'Figures' / 'group_classification.svg'))
```
